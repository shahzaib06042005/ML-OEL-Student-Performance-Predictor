```python
#importing libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report, mean_squared_error
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('student-mat.csv', sep=';')
df.head()
```

```python
# 3.1Check for missing values
print(df.isnull().sum())

# Encode categorical variables
df = pd.get_dummies(df, drop_first=True)
```

```python
#3.2 Features and target
X = df[['studytime', 'absences', 'G1', 'G2']]
y_marks = df['G3']  # final marks

# Define pass/fail threshold
df['pass_fail'] = np.where(df['G3'] >= 10, 1, 0)
y_class = df['pass_fail']
```

```python
# Train - test splitting
X_train, X_test, y_train_marks, y_test_marks = train_test_split(X, y_marks, test_size=0.2, random_state=42)
_, _, y_train_class, y_test_class = train_test_split(X, y_class, test_size=0.2, random_state=42)
```

```python
#Fearture Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#Model-Training using linear regression
lin_reg = LinearRegression()
lin_reg.fit(X_train_scaled, y_train_marks)

y_pred_marks = lin_reg.predict(X_test_scaled)
print("Linear Regression RMSE:", np.sqrt(mean_squared_error(y_test_marks, y_pred_marks)))
```

```python
#Model-Training using logistic regression(Pass/fail batayega)
log_reg = LogisticRegression()

log_reg.fit(X_train_scaled, y_train_class)

y_pred_log = log_reg.predict(X_test_scaled)
```

```python
#Model-Training using Decision Tree(Pass/fail batayega)
tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_train_scaled, y_train_class)
y_pred_tree = tree_clf.predict(X_test_scaled)
```

```python
#Comparison
print("=== Logistic Regression Performance ===")
print("Accuracy:", accuracy_score(y_test_class, y_pred_log))
print("Precision:", precision_score(y_test_class, y_pred_log))
print("Recall:", recall_score(y_test_class, y_pred_log))
print(classification_report(y_test_class, y_pred_log))

print("\n=== Decision Tree Performance ===")
print("Accuracy:", accuracy_score(y_test_class, y_pred_tree))
print("Precision:", precision_score(y_test_class, y_pred_tree))
print("Recall:", recall_score(y_test_class, y_pred_tree))
print(classification_report(y_test_class, y_pred_tree))
```
